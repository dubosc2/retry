```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
month = np.array(["Jan","Feb","Mar"])
productA = np.array([120,130,125])
productB = np.array([150,160,170])
productC = np.array([100,110,105])
data = {"Month":["Jan","Feb","Mar"],"productA":[120,130,125],"productB":[150,160,170],"productC":[100,110,105]}
df = pd.DataFrame(data)
print(df)
```

```python
df.describe()
```

# Here mean is nothing but its a average of that perticular product like productA 125.0, productB 160.0 and productC 105.0 is taverage value of the product.
# Here min is nothing but its a min value (smallest number) of that perticular product like productA 120, productB 150 and productC 100 is min value of the product.
# Here count is nothing but its shows how many items of that perticular product like productA 3, productB 3 and productC 3 is count of the product.

```python
month = ["Jan","Feb","Mar"]
productA = [120,130,125]
productB = [150,160,170]
productC = [100,110,105]
plt.plot(month, productA,color="Black",linewidth=2,label="Product A")
plt.plot(month, productB,color="DarkGreen",linewidth=2,linestyle="--",label="Product B")
plt.plot(month, productC,color="Red",linewidth=2,linestyle="-",label="Product C")
plt.xlabel("MONTH")
plt.ylabel("PRODUCT")
plt.legend(loc="best",frameon=False)
plt.tight_layout()
plt.show()
```

# Chart Explanation
Here its a very simple to undrstand on line chart, 'x' axis is months and 'y' axis is product. line represent in chart product A and C sales is increase in month of Feb but next month in March is down. product B sales is increase in every months its represent upgrowing product.

```python
# find median using numpy
median_productA = np.median(productA)
median_productB = np.median(productB)
median_productC = np.median(productC)
print("Median of Product A:", median_productA)
print("Median of Product B:", median_productB)
print("Median of Product C:", median_productC)
```

```python
# barchart or histogram using matplotlib
labels = ['Product A', 'Product B', 'Product C']
values = [np.mean(productA), np.mean(productB), np.mean(productC)]
plt.bar(labels, values, color=['black', 'darkgreen', 'red'])
plt.xlabel('Products')
plt.ylabel('Average Sales')
plt.title('Average Sales of Products')
plt.show()
```

Used numpy to add statistics on median using Numpy. Added barchart to show average sales data.
