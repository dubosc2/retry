```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
   
Month = np.array(["Jan","Feb","Mar"])
print(Month)

ProductA = np.array([120,130,125])
print(ProductA)

ProductB = np.array([150,160,170])
print(ProductB)

ProductC = np.array([100,110,105])
print(ProductC)
```

```python
#create dataframe
df = pd.DataFrame({'Month': ['January', 'February', 'March'], 'ProductA': [120, 130, 125], 'ProductB': [150, 160, 170], 'ProductC': [100,110,105]})

print(df)
```

```python
df.describe()
```

- Mean: The sum of values divided by the number of values.
- Min: The smallest value observed.
- Count: The number of entries in a column.

```python
#plot line graph using matplotlib

plt.xlabel('Month')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.ylabel('Sales')
plt.title('Monthly Sales of Product A, B, and C')
plt.plot(df['Month'], df['ProductA'], label='Product A', marker='o')
plt.plot(df['Month'], df['ProductB'], label='Product B', marker='s')
plt.plot(df['Month'], df['ProductC'], label='Product C', marker='^')
plt.legend()
plt.show()
```

The chart shows that Product B has an increase in sales from February to March while the other products show a decrease.
